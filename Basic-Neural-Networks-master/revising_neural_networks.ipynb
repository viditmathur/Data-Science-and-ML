```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import sklearn
from sklearn.preprocessing import OneHotEncoder
from sklearn.datasets import make_circles
import keras
from keras.utils import np_utils
from keras.layers import Dense,Activation
from keras.models import Sequential
```

```python
#making circles where X is the shape and Y is the label category i.e 0 and 1
X,y = make_circles(n_samples=1000,noise=0.1,factor=0.1)

print(X.shape,y.shape)
```

```python
Y = np_utils.to_categorical(y)
print(Y.shape)
```

```python
#for label 0 make red circles and for 1 make blue circles
for ix in range(X.shape[0]):
    if y[ix]==0:
        
        plt.scatter(X[ix,0],X[ix,1],color="red")
    else:
        plt.scatter(X[ix,0],X[ix,1],color="blue")
```

```python
model = Sequential()

model.add(Dense(8,input_shape=(2,)))
model.add(Activation('relu'))

model.add(Dense(2))
model.add(Activation('relu'))

model.summary()
model.compile(loss='categorical_crossentropy',optimizer='nadam',metrics=['accuracy'])
```

```python
split = int(0.8*X.shape[0])

X_train = X[:split]
y_train = Y[:split]

X_test = X[split:]
y_test = Y[split:]

print(X_train.shape,y_train.shape)
print(X_test.shape,y_test.shape)
```

```python
hist = model.fit(X_train,y_train,
                epochs=200,
                shuffle=True,
                batch_size=100,
                validation_data=(X_test,y_test))
```

```python
data =hist.history
plt.figure(0)
plt.plot(data['loss'],'g')
plt.plot(data['val_loss'],'r')

plt.figure(1)
plt.plot(data['acc'],'g')
plt.plot(data['val_acc'],'r')

plt.show()
```
